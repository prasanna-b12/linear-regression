# Linear Regression | Basics - Machine Learning Model

Linear regression is a linear relationship between the input variable (x) and the single output variable (y). It is used when we want to predict the value of a variable based on the value of another variable. The variable we want to predict is called the <b>dependent variable</b> (or sometimes, the outcome variable).
- X, <u>Independent Variable</u> otherwise called predictors variable
- Y, <u>Dependent Variable</u> otherwise called predicted variable

<b>For example,</b> Assume we've the dataset which includes study Time and the Grade of the Students in a class. Here...
- Study time - Independent Variable
- Grade - Dependent Variable

More Study time increases marks(Grade) that's why we call grade as the dependent variable.

#### Formula to calculate the simple linear regression is shown below

<h2>Y' = Mx + c</h2>
<h4>M (slope) = r (S <sub>y</sub> / S <sub>x</sub>)

C (intercept) = ȳ - M x̄</h4>

# Calculation of Regression line

<b>- Importing the Libraries</b>

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

<b>- Importing the dataset</b>

```python
df = pd.read_csv("csv/students.csv")
df.head()
```

<b>- Screening the dataset</b>

```python
#note that .values is the attribute which converts it to numpy array
x = df.iloc[:,0].values.reshape(-1,1)
y = df.iloc[:,1]

plt.scatter(x,y)
plt.xlabel('No. of Hours study')
plt.ylabel('Grade')
plt.show()
```

```python
#calculating the regression line
reg = LinearRegression().fit(x,y)

print("Predicted Values:\n", reg.predict(x))

plt.scatter(x,y)
plt.plot(x,reg.predict(x), color='red')
plt.title("Regression Line")
plt.xlabel('No. of Hours study')
plt.ylabel('Grade')
plt.show()

print("Slope Value =", reg.coef_[0],"\n")

print("Intercept Value =", reg.intercept_,"\n")

print("Correlatin Value =", df['study_time'].corr(df['grade']),"\n")

print("R-Squared Value =", reg.score(x,y),"\n")
```

# Creating the Linear Regression Model using Sklearn

<b>- Splitting the dataset</b>

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.7, test_size = 0.3)
```

- <b>train_size</b> is the number that defines the size of the training set. If you provide a float, then it must be between 0.0 and 1.0 and will define the share of the dataset used for testing. If you provide an int, then it will represent the total number of the training samples. The default value is None.
 
- <b>test_size</b> is the number that defines the size of the test set. It’s very similar to train_size. You should provide either train_size or test_size. If neither is given, then the default share of the dataset that will be used for testing is 0.25, or 25 percent.

- <b>random_state</b> is the object that controls randomization during splitting. It can be either an int or an instance of RandomState. The default value is None.

- <b>shuffle</b> is the Boolean object (True by default) that determines whether to shuffle the dataset before applying the split.

Pictorial representation for spiltting the dataset is shown below...

<b>As you can see the train dataset, it randomly select the X and Y values for train as well as test..</b>

```python
plt.scatter(x_train,y_train, color='green',)
plt.scatter(x_test,y_test, color='red')
plt.legend(["Training","Testing"], loc='lower right')
plt.title("Train Test Split")
plt.xlabel('No. of Hours study')
plt.ylabel('Grade')
plt.show()
```

<b>- Creating the Regression Model</b>

```python
#assign the variable fit the model with the train dataset
model = LinearRegression().fit(x_train,y_train)

print("Slope Value =", model.coef_[0])

print("Intercept Value =", model.intercept_)
```

<b>- Now the model is created, we can use the model to predict</b>

```python
print("Predicted Values:\n", model.predict(x_test),"\n")

print("Accuracy Score for Training", model.score(x_train,y_train),"\n")

print("Accuracy Score for Testing", model.score(x_test,y_test))
```
